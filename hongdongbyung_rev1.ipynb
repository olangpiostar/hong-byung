```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# data proccessing
data = pd.read_csv('./data/train.csv')
Name = data.Name
test1 = list(map(lambda x:x.split('.')[0].split(' ')[-1],list(data.Name)))
data['Name']=test1
# changind Name
Name_number = []
for i in data['Name']:
    if i =="Mr":
        Name_number.append(1)
    elif i =="Mrs":
        Name_number.append(2)
    elif i =="Miss":
        Name_number.append(3)
    elif i =="Rev":
        Name_number.append(4)
    elif i == "Dr":
        Name_number.append(5)
    else:
        Name_number.append(6)
data['Name'] = Name_number

# changing sex
sex_number = []
for i in data['Sex']:
    if i =='male':
        sex_number.append(1)
    else:
        sex_number.append(2)
data['Sex'] = sex_number

# changing embarked
embarked_number = []
for i in data['Embarked']:
    if i == 'C':
        embarked_number.append(1)
    elif i == 'S':
        embarked_number.append(2)
    elif i == 'Q':
        embarked_number.append(3)
    else:
        embarked_number.append(4)
data['Embarked']=embarked_number

# control missing age data
data['Age']= data['Age'].fillna(0)

#train data set
y = data['Survived']
x = data.drop('Survived',axis = 1).drop('Ticket',axis = 1).drop('Cabin',axis =1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 10)
```

```python
rf = RandomForestClassifier(random_state = 0)
rf.fit(x_train,y_train)
```

```python
model_predict=rf.predict(x_test)
print("Accuracy : {}".format(accuracy_score(y_test,model_predict)))
```
